```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.linalg as la
```

```python
data = pd.read_csv('CanadianFoodPricesJuly_November_2019_clean.csv', names=['Jul','Aug','Sep','Oct','Nov'])
#pd.options.display.max_rows = 1000
data.apply(pd.DataFrame.describe, axis=1)
gas_price = data.iloc[-1]


N = 4
x = [0,1,2,3,4]
y = np.zeros(N+1)

for i in range(1,N+1):
    y[i] = ((gas_price[i]/gas_price[i-1])-1)
X = np.vander(x,increasing=True)
a = la.solve(X,y)
xs = np.linspace(0,N,153)
ys = sum([a[k]*xs**k for k in range(0,N+1)])
```
